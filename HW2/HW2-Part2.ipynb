# HW 2 - K-fold Cross Validation

In this notebook I have tested the winning archetecture of the previous part on CIFAR10 dataset with K-fold cross validation technique. The following parts include loading dataset and data preparation and running the k-fold cross validation on the CNN with k = 5. The last part visualize the average accuracy and loss of the 5-folds per epoch.

```python
import numpy as np
import keras
from keras.models import Sequential
from keras import layers
from keras.optimizers import SGD , adam
from keras.datasets import cifar10
from keras.utils.np_utils import to_categorical  
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Activation, Dropout
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
# load data
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()
```

```python
# initial parameters
class_names = ['airplan', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
data_size = train_images.shape[0]
batch_size = 32
epochs = 100
k = 5
valid_size = data_size / k
train_size = data_size - valid_size
train_acc = []
train_loss = []
val_acc = []
val_loss = []
```

```python
# preparing dataset
train_labels = to_categorical(train_labels, num_classes = len(class_names))
test_labels = to_categorical(test_labels, num_classes = len(class_names))
test_images = test_images.astype('float32') / 255.0
train_images = train_images.astype('float32') / 255.0
```

```python
# creating validation set
shuffled_indices = np.random.permutation(data_size)
train_set_x = train_images[shuffled_indices]
train_set_y = train_labels[shuffled_indices]
```

```python
def kfold_cross_validation(iter):

    validation_set_x = train_set_x[(iter-1) * valid_size : iter * valid_size]
    validation_set_y = train_set_y[(iter-1) * valid_size : iter * valid_size]

    train_set_x_new = np.delete(train_set_x, [range((iter-1) * valid_size , iter * valid_size)], axis = 0)
    train_set_y_new = np.delete(train_set_y, [range((iter-1) * valid_size , iter * valid_size)], axis = 0)

    return train_set_x_new, train_set_y_new, validation_set_x, validation_set_y
```

```python
for iter in range(1,k+1):
    # creating model
    model = Sequential()
    model.add(Conv2D(32, (3, 3), activation = 'relu', input_shape = (32, 32, 3)))
    model.add(Conv2D(32, (3, 3), activation = 'relu'))
    model.add(Conv2D(64, (3, 3), activation = 'relu', strides = 2))
    model.add(Dropout(0.5))
    model.add(Conv2D(64, (3, 3), activation = 'relu'))
    model.add(Conv2D(128, (3, 3), activation = 'relu', strides=2))
    model.add(Conv2D(128, (3, 3), activation = 'relu'))
    model.add(Flatten())
    model.add(Dropout(0.5))
    model.add(Dense(64, activation = 'relu'))
    model.add(layers.normalization.BatchNormalization())
    model.add(Dense(10, activation = 'softmax'))

    model.compile(loss = 'categorical_crossentropy', metrics = ['accuracy'], optimizer = 'rmsprop')

    train_x, train_y, validation_x, validation_y = kfold_cross_validation(iter)
    
    # Data Augmentation
    train_datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )

    validation_datagen = ImageDataGenerator()

    train_generator = train_datagen.flow(train_x, train_y, batch_size=32)
    validation_generator = validation_datagen.flow(validation_x, validation_y, batch_size=32)
    
    # training model
    hist = model.fit_generator(train_generator,validation_data = validation_generator, 
                        validation_steps=len(train_set_x) / 32,
                        steps_per_epoch=len(train_set_x) / 32,
                        epochs = epochs, verbose=2)
    

    train_acc.append(hist.history['acc'])
    val_acc.append(hist.history['val_acc'])
    train_loss.append(hist.history['loss'])
    val_loss.append(hist.history['val_loss'])
```

```python
#Visualize results
epoch_count = range(1, epochs + 1) 

plt.plot(epoch_count, np.mean(train_acc, axis = 0), 'r--')
plt.plot(epoch_count, np.mean(val_acc, axis = 0), 'b-')
plt.legend(['Training acc', 'Test acc'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
