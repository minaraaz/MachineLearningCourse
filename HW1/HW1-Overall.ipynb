```python
from __future__ import print_function
import keras
import numpy as np
from keras.datasets import mnist
from keras import models
from keras import layers
from keras.utils import to_categorical
import matplotlib.pyplot as plt
from copy import deepcopy
```

```python
# data load
(train_images_original, train_labels_original), (test_images_original, test_labels_original) = mnist.load_data()
```

```python
# data reshape and black and white
train_images = train_images_original.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255.0

test_images = test_images_original.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255.0
```

```python
#Converting Labels to One Hot Encoding
train_labels = (np.arange(np.max(train_labels_original) + 1) == train_labels_original[:, None]).astype(float)
test_labels = (np.arange(np.max(test_labels_original) + 1) == test_labels_original[:, None]).astype(float)
```

```python
# initial variables
batch_size = 600
epochs = 20
learning_rate = 1
m = 10
num_classes = 10
img_rows, img_cols = 28, 28
data_size = train_images.shape[0]
test_size = 10000
dimension = train_images.shape[1]
digit_classifiers_1 = []
digit_classifiers_2 = []
digit_classifier_3 = []
models_train_accuracy_1 = []
models_test_accuracy_1 = []
models_train_accuracy_2 = []
models_test_accuracy_2 = []
models_train_accuracy_3 = []
models_test_accuracy_3 = []
predicted_labels_train_1 = np.zeros((1,train_labels_original.shape[0]))
predicted_labels_test_1 = np.zeros((1,test_labels_original.shape[0]))
predicted_labels_train_2 = np.zeros((1,train_labels_original.shape[0]))
predicted_labels_test_2 = np.zeros((1,test_labels_original.shape[0]))
predicted_labels_train_3 = np.zeros((1,train_labels_original.shape[0]))
predicted_labels_test_3 = np.zeros((1,test_labels_original.shape[0]))
```

```python
# classifier with squared error loss function
class classifier1:
    
    def __init__(self, number, batch_size, dimension, learning_rate,train_labels_original,test_labels_original):
        self.number=number
        self.weight = np.zeros((dimension, 1))
        self.bias = 0
        self.dimension = dimension
        self.learning_rate = learning_rate
        self.batch_size = batch_size
        self.Label_train_binary = [1 if a == number else 0 for a in train_labels_original]
        self.Label_test_binary = [1 if a == number else 0 for a in test_labels_original]
        self.Label_train_binary = np.asarray(self.Label_train_binary)
        self.Label_test_binary = np.asarray(self.Label_test_binary)
        self.train_labels_shuffled = np.zeros((1,train_labels_original.shape[0]))
    
    def sigmoid(self, z):
        s = 1.0/(1.0 + np.exp(-z))
        return s
    
    def backward(self, X, Y):
        z = np.dot(self.weight.T,X) + self.bias
        A = self.sigmoid(z)

        gradient_weight = 1.0/self.batch_size * np.dot(X, ((A-Y)*A*(1-A)).T)
        gradient_bias = 1.0/self.batch_size * np.sum(((A-Y)*A*(1-A)))

        return gradient_weight, gradient_bias
    
    def update(self, X, Y):
        X = X.transpose()
        gradient_weight, gradient_bias = self.backward(X, Y)

        self.weight = self.weight - self.learning_rate * gradient_weight
        self.bias = self.bias - self.learning_rate * gradient_bias


    def predict (self, X):
        Y_prediction = np.zeros((1,X.shape[0]))
        X = X.transpose()

        A = self.sigmoid(np.dot(self.weight.T, X) + self.bias)
        
        return A
    
    def Label_shuffle(self, shuffled_indices):
        self.train_labels_shuffled = self.Label_train_binary[shuffled_indices]

    def train(self, i, xi):
        yi = self.train_labels_shuffled[i : i + self.batch_size]
        self.update(xi,yi)
```

```python
# classifier with binary cross entropy loss function
class classifier2:
    
    def __init__(self, number, batch_size, dimension, learning_rate,train_labels_original,test_labels_original):
        self.number=number
        self.weight = np.zeros((dimension, 1))
        self.bias = 0
        self.dimension = dimension
        self.learning_rate = learning_rate
        self.batch_size = batch_size
        self.Label_train_binary = [1 if a == number else 0 for a in train_labels_original]
        self.Label_test_binary = [1 if a == number else 0 for a in test_labels_original]
        self.Label_train_binary = np.asarray(self.Label_train_binary)
        self.Label_test_binary = np.asarray(self.Label_test_binary)
        self.train_labels_shuffled = np.zeros((1,train_labels_original.shape[0]))
    
    def sigmoid(self, z):
        s = 1.0/(1.0 + np.exp(-z))
        return s
    
    def backward(self, X, Y):
        z = np.dot(self.weight.T,X) + self.bias
        A = self.sigmoid(z)

        gradient_weight = 1.0/self.batch_size * np.dot(X, (A-Y).T)
        gradient_bias = 1.0/self.batch_size * np.sum(A-Y)

        return gradient_weight, gradient_bias
    
    def update(self, X, Y):
        X = X.transpose()
        gradient_weight, gradient_bias = self.backward(X, Y)

        self.weight = self.weight - self.learning_rate * gradient_weight
        self.bias = self.bias - self.learning_rate * gradient_bias


    def predict (self, X):
        Y_prediction = np.zeros((1,X.shape[0]))
        X = X.transpose()

        A = self.sigmoid(np.dot(self.weight.T, X) + self.bias)
        
        return A

    def Label_shuffle(self, shuffled_indices):
        self.train_labels_shuffled = self.Label_train_binary[shuffled_indices]

    def train(self, i, xi):
        yi = self.train_labels_shuffled[i : i + self.batch_size]
        self.update(xi,yi)
```

```python
#softmax classifier with categorical cross entropy
class softmax_classifier:
    
    def __init__(self, batch_size, dimension, m, learning_rate):
        self.weight = np.zeros((dimension, m))
        self.bias = np.zeros((m, 1))
        self.dimension = dimension
        self.learning_rate = learning_rate
        self.batch_size = batch_size
        self.num_class = m
    
    def softmax(self, z):
        z = z - np.max(z, axis=0)
        s = np.exp(z)/np.sum(np.exp(z), axis=0)
        return s

    def backward(self, X, Y):
        z = np.dot(self.weight.T,X) + self.bias
        A = self.softmax(z)
        A = np.repeat(A[np.newaxis], self.num_class, axis = 0) - np.repeat(np.identity(self.num_class)[:,:, np.newaxis], self.batch_size, axis=2)
        YA = np.repeat(Y.T[:, np.newaxis], self.num_class, axis = 1) * A
        YA = np.sum(YA, axis=0)

        gradient_weight = 1.0/self.batch_size * np.dot(X, YA.T)
        gradient_bias = 1.0/self.batch_size * np.sum(YA, axis=1,keepdims=True)

        return gradient_weight, gradient_bias
    
    def update(self, X, Y):
        X = X.transpose()
        gradient_weight, gradient_bias = self.backward(X, Y)

        self.weight = self.weight - self.learning_rate * gradient_weight
        self.bias = self.bias - self.learning_rate * gradient_bias


    def predict (self, X):
        X = X.transpose()
        A = self.softmax(np.dot(self.weight.T, X) + self.bias)
        
        return A
```

```python
# creating 10 classifiers for squared error loss function
for digit in range(10):
        digit_classifiers_1.append(classifier1(digit, batch_size, dimension, learning_rate,train_labels_original,test_labels_original))
```

```python
# creating 10 classifiers for binary cross entropy loss function
for digit in range(10):
        digit_classifiers_2.append(classifier2(digit, batch_size, dimension, learning_rate,train_labels_original,test_labels_original))
```

```python
# creating softmax classifier
digit_classifier_3 = softmax_classifier(batch_size, dimension, m, learning_rate)
```

```python
# training classifiers
for epoch in range(epochs):
        Y_train_dic_1 = []
        Y_test_dic_1 = []
        Y_train_dic_2 = []
        Y_test_dic_2 = []
        Y_train_dic_3 = []
        Y_test_dic_3 = []
        
        shuffled_indices = np.random.permutation(data_size)
        train_images_shuffled = train_images[shuffled_indices]
        train_labels_shuffled = train_labels[shuffled_indices]
        
        for digit in range(10):
                digit_classifiers_1[digit].Label_shuffle(shuffled_indices)
                digit_classifiers_2[digit].Label_shuffle(shuffled_indices)
        for i in range(0, data_size, batch_size):
                xi = train_images_shuffled[i : i + batch_size]
                yi = train_labels_shuffled[i : i + batch_size]
                
                for digit in range(10):
                        digit_classifiers_1[digit].train(i,xi)
                        digit_classifiers_2[digit].train(i,xi)
                        digit_classifier_3.update(xi,yi)
        
        for digit in range(10):
                Y_train_dic_1.append(digit_classifiers_1[digit].predict(train_images))
                Y_test_dic_1.append(digit_classifiers_1[digit].predict(test_images))
                
                Y_train_dic_2.append(digit_classifiers_2[digit].predict(train_images))
                Y_test_dic_2.append(digit_classifiers_2[digit].predict(test_images))

        predicted_labels_train_1 = np.squeeze(np.argmax(Y_train_dic_1, axis = 0))
        predicted_labels_test_1 = np.squeeze(np.argmax(Y_test_dic_1, axis = 0))
        
        predicted_labels_train_2 = np.squeeze(np.argmax(Y_train_dic_2, axis = 0))
        predicted_labels_test_2 = np.squeeze(np.argmax(Y_test_dic_2, axis = 0))
        
        predicted_labels_train_3 = np.squeeze(np.argmax(digit_classifier_3.predict(train_images), axis = 0))
        predicted_labels_test_3 = np.squeeze(np.argmax(digit_classifier_3.predict(test_images), axis = 0))

        train_accuracy_1 = np.mean([int(i==j) for i, j in zip(predicted_labels_train_1, train_labels_original)]) * 100.0
        models_train_accuracy_1.append(train_accuracy_1)
        
        test_accuracy_1 = np.mean([int(i==j) for i, j in zip(predicted_labels_test_1, test_labels_original)]) * 100.0
        models_test_accuracy_1.append(test_accuracy_1)
        
        train_accuracy_2 = np.mean([int(i==j) for i, j in zip(predicted_labels_train_2, train_labels_original)]) * 100.0
        models_train_accuracy_2.append(train_accuracy_2)
        
        test_accuracy_2 = np.mean([int(i==j) for i, j in zip(predicted_labels_test_2, test_labels_original)]) * 100.0
        models_test_accuracy_2.append(test_accuracy_2)
        
        train_accuracy_3 = np.mean([int(i==j) for i, j in zip(predicted_labels_train_3, train_labels_original)]) * 100.0
        models_train_accuracy_3.append(train_accuracy_3)
        
        test_accuracy_3 = np.mean([int(i==j) for i, j in zip(predicted_labels_test_3, test_labels_original)]) * 100.0
        models_test_accuracy_3.append(test_accuracy_3)
        
        print ("epoch " + str(epoch))
        print ("Classifier1 ==> train accuracy: {0:0.4f} ==> test accuracy: {1:0.4f}".format(train_accuracy_1, test_accuracy_1)) 
        print ("Classifier2 ==> train accuracy: {0:0.4f} ==> test accuracy: {1:0.4f}".format(train_accuracy_2, test_accuracy_2)) 
        print ("Classifier3 ==> train accuracy: {0:0.4f} ==> test accuracy: {1:0.4f}".format(train_accuracy_3, test_accuracy_3))
```

```python
network = models.Sequential()
#network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network.add(layers.Dense(10, activation='softmax'))

network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])


history = network.fit(train_images, 
                      train_labels, 
                      epochs=epochs, 
                      batch_size=128, 
                      validation_data=(test_images, test_labels))

score = network.evaluate(test_images, test_labels, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
def return_next(visited, stack, counter):
    dim = len(visited)
    for col in range(dim):
        for row in range(dim):
            if not visited[row][col]:
                stack.append((row, col))
                counter += 1
                return stack, counter

def CC_counter_four(matrix):
    visited = deepcopy(matrix)
    dim = len(matrix)
    for col in range(dim):
        for row in range(dim):
            if matrix[row][col] == 0:
                visited[row][col] = False
            else:
                visited[row][col] = True
    
    stack = []
    counter = 0

    while not all(v for row in visited for v in row):
        if not stack:
            stack, counter = return_next(visited, stack, counter)
            
        popped = stack.pop()
        visited[popped[0]][popped[1]] = True
        if popped[0] - 1 >= 0 and not visited[popped[0] - 1][popped[1]]:
            stack.append((popped[0] - 1, popped[1]))
        if popped[0] + 1 < dim and not visited[popped[0] + 1][popped[1]]:
            stack.append((popped[0] + 1, popped[1]))
        if popped[1] - 1 >= 0 and not visited[popped[0]][popped[1] - 1]:
            stack.append((popped[0], popped[1] - 1))
        if popped[1] + 1 < dim and not visited[popped[0]][popped[1] + 1]:
            stack.append((popped[0], popped[1] + 1))
                
    return counter

def CC_counter_eight(matrix):
    visited = deepcopy(matrix)
    dim = len(matrix)
    for col in range(dim):
        for row in range(dim):
            if matrix[row][col] == 0:
                visited[row][col] = False
            else:
                visited[row][col] = True
    stack = []
    counter = 0

    while not all(v for row in visited for v in row):
        if not stack:
            stack, counter = return_next(visited, stack, counter)
            
        popped = stack.pop()
        visited[popped[0]][popped[1]] = True
        if popped[0] - 1 >= 0 and not visited[popped[0] - 1][popped[1]]:
            stack.append((popped[0] - 1, popped[1]))
        if popped[0] + 1 < dim and not visited[popped[0] + 1][popped[1]]:
            stack.append((popped[0] + 1, popped[1]))
        if popped[1] - 1 >= 0 and not visited[popped[0]][popped[1] - 1]:
            stack.append((popped[0], popped[1] - 1))
        if popped[1] + 1 < dim and not visited[popped[0]][popped[1] + 1]:
            stack.append((popped[0], popped[1] + 1))
        if popped[0] + 1 < dim and popped[1] + 1 < dim and not visited[popped[0] + 1][popped[1] + 1]:
            stack.append((popped[0] + 1,popped[1] + 1))
        if popped[0] - 1 >= 0 and popped[1] - 1 >= 0 and not visited[popped[0] - 1][popped[1] - 1]:
            stack.append((popped[0] - 1,popped[1] - 1))
        if popped[0] - 1 >= 0 and popped[1] + 1 < dim and not visited[popped[0] - 1][popped[1] + 1]:
            stack.append((popped[0] - 1,popped[1] + 1))
        if popped[0] + 1 < dim and popped[1] - 1 >= 0 and not visited[popped[0] + 1][popped[1] - 1]:
            stack.append((popped[0] + 1,popped[1] - 1))
    
                
    return counter

def Value_summation(matrix):
        return sum(map(sum, matrix))
```

```python
# adding new feature
for i in range(data_size):
    np.append(train_images[i],CC_counter_four(train_images_original[i].astype('float32') / 255.0)/3.0)
    np.append(train_images[i],CC_counter_eight(train_images_original[i].astype('float32') / 255.0)/3.0)

    if i < test_size:
        np.append(test_images[i],CC_counter_four(test_images_original[i].astype('float32') / 255.0)/3.0)
        np.append(test_images[i],CC_counter_eight(test_images_original[i].astype('float32') / 255.0)/3.0)
```
